```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/list/ls091520106/')
```

```python
#page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
#print(soup.prettify())
```

```python
Movie_name=[]
Rating=[]
Release_date=[]
```

```python
movie=soup.find_all('h3', class_='lister-item-header')
movie[0:4]
```

```python
for i in movie:
    
    Movie_name.append(i.get_text().replace("\n",""))
Movie_name
```

```python
Ratings=soup.find_all('div', class_='ipl-rating-star small')
Ratings
```

```python
for i in Ratings:
    Rating.append(i.get_text().replace("\n",""))
Rating
```
