4) Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://bookpage.com/reviews')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
Book_name=[]
Author_name=[]
genre=[]
book_review=[]
```

```python
bname=soup.find_all('h4',class_='italic')
bname=bname[:5]
```

```python
for i in bname:
    Book_name.append(i.get_text().replace("\n",""))
Book_name
```

```python
aname=soup.find_all('p',class_='sans bold')
aname=aname[:5]
```

```python
for i in aname:
    Author_name.append(i.get_text().replace("\n",""))
Author_name
```

```python
genre_=soup.find_all('p',class_='genre-links hidden-phone')
genre_=genre_[:5]
```

```python
for i in genre_:
    genre.append(i.get_text().replace("\n",""))
genre
```

```python
bookreview=soup.find_all('p',class_='excerpt')
bookreview=bookreview[:5]
```

```python
book_review
for i in bookreview:
    book_review.append(i.get_text().replace("\n",""))
book_review
```

```python
print(len(Book_name),len(Author_name),len(genre),len(book_review))
```

```python
import pandas as pd
books=pd.DataFrame({})
books['Book_name']=Book_name
books['Author_name']=Author_name
books['Genre']=genre
books['Book_review']=book_review
```
